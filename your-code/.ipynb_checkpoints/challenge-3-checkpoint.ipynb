```python
test = orders[[ 'CustomerID', 'amount_spent']].groupby(['CustomerID']).agg({'amount_spent':'sum'})  #Te da un Dataframe anidado
#test = orders[[ 'CustomerID', 'amount_spent']].groupby(['CustomerID'])['amount_spent'].sum()    #######Te da una serie de Panda
test = test.add_prefix('Total_').reset_index()
#np.quantile(test['Total_amount_spent'], .50)
qua_95 = test['Total_amount_spent'].quantile(.95)
qua_75 = test['Total_amount_spent'].quantile(.75)
test['Customer Type'] = 'Customer'
test.loc[test['Total_amount_spent']>=qua_75, 'Customer Type'] = 'Prefered Customer'
test.loc[test['Total_amount_spent']>=qua_95, 'Customer Type'] = 'VIP Customer'
test


```

Provide your solution for Q2 below:

```python
test2 = (orders.merge(test, left_on='CustomerID', right_on='CustomerID'))[['Country', 'Customer Type']]
test2.groupby(['Country', 'Customer Type']).agg({'Customer Type':'count'})
```
